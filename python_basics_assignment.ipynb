# ***1. ENCAPSULATION***

***Example 1***

```python
class AudioFile:
    def __init__(self, name, duration):
        self.__name = name          # private
        self.__duration = duration  # private

    def get_name(self):
        return self.__name

    def get_duration(self):
        return self.__duration
```

***Example 2***

```python
class Recorder:
    def __init__(self):
        self.__is_recording = False

    def start(self):
        self.__is_recording = True

    def stop(self):
        self.__is_recording = False

    def is_recording(self):
        return self.__is_recording
```

***Example 3***

```python
class VolumeControl:
    def __init__(self):
        self.__level = 50  # private volume level

    def set_volume(self, value):
        self.__level = value

    def get_volume(self):
        return self.__level
```

#***2. Inheritance***

```python
class Animal:
    def speak(self):
        return "Some sound"

class Dog(Animal):
    pass   # inherits speak()

class Cat(Animal):
    pass   # inherits speak()

class Cow(Animal):
    pass   # inherits speak()

print(Dog().speak())
print(Cat().speak())
print(Cow().speak())
```

#***3. Polymorphism***

```python
class Shape:
    def area(self):
        return 0
class Square(Shape):
    def area(self):
        return "Area = side × side"
class Rectangle(Shape):
    def area(self):
        return "Area = length × width"
class Circle(Shape):
    def area(self):
        return "Area = π × r²"

shapes = [Square(), Rectangle(), Circle()]
for s in shapes:
    print(s.area())
```

#***4. Extraction***

```python
class DataExtractor:
    def extract(self, data):
        return f"Extracting data: {data}"
class NameExtractor:
    def extract(self, student_name):
        return f"Extracted name: {student_name}"
class PriceExtractor:
    def extract(self, price):
        return f"Extracted price: ₹{price}"

print(NameExtractor().extract("Yugavardhan"))
```
